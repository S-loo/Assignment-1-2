## Kaggle Dataset: International Football Results (1872–2024)

This analysis follows the structured exercise to explore international football trends using the Kaggle dataset.

### Step 1: Load the CSV
First, we load the dataset and our custom analytics module.

```python
import football_analytics as fa
import pandas as pd

df = fa.load_data()
df.head()
```

### Basic Exploration
We start with a fundamental overview of the dataset scope.

**Questions:**
1. How many matches are in the dataset?
2. What is the earliest and latest year in the data?
3. How many unique countries are there?
4. Which team appears most frequently as home team?

```python
stats = fa.get_basic_stats(df)

print(f"Total Matches Analyzed: {stats['total_matches']:,}")
print(f"Temporal Scope: {stats['earliest_year']} to {stats['latest_year']}")
print(f"Unique Countries Represented: {stats['unique_countries']}")
print(f"Most Active Home Team: {stats['most_freq_home']} ({stats['most_freq_home_count']} matches)")
```

### Goals Analysis
Analyzing goal patterns reveals how the game has been played over decades.

**Questions:**
5. What is the average number of goals per match?
6. What is the highest scoring match?
7. Are more goals scored at home or away?
8. What is the most common total goals value?

```python
goals = fa.analyze_goals(df)
high_score = goals['highest_scoring_match']

print(f"Average Goals per Match: {goals['avg_goals']:.2f}")
print(f"Most Common Match Score (Total Goals): {goals['most_common_total']}")
print(f"Highest Scoring Match: {high_score['home']} vs {high_score['away']} ({high_score['goals']} goals)")
```

### Match Results
We investigate outcome distributions to quantify concepts like 'Home Advantage'.

**Questions:**
9. What percentage of matches are home wins?
10. Does home advantage exist?
11. Which country has the most wins historically?

```python
outcomes = fa.get_match_outcomes(df)

for result, pct in outcomes.items():
    print(f"{result}: {pct:.1f}%")
```
